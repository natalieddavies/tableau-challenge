```python
# DEPENDENCIES
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from datetime import datetime

# IMPORT DATA FILES
p01 = "JC-202001-citibike-tripdata.csv"
p02 = "JC-202002-citibike-tripdata.csv"
p03 = "JC-202003-citibike-tripdata.csv"
p04 = "JC-202004-citibike-tripdata.csv"
p05 = "JC-202005-citibike-tripdata.csv"
p06 = "JC-202006-citibike-tripdata.csv"
p07 = "JC-202007-citibike-tripdata.csv"
p08 = "JC-202008-citibike-tripdata.csv"
p09 = "JC-202009-citibike-tripdata.csv"
```

```python
# READ IN DATA
jan = pd.read_csv(p01)
feb = pd.read_csv(p02)
mar = pd.read_csv(p03)
apr = pd.read_csv(p04)
may = pd.read_csv(p05)
jun = pd.read_csv(p06)
jul = pd.read_csv(p07)
aug = pd.read_csv(p08)
sep = pd.read_csv(p09)
```

```python
# PREVIEW ONE OF THE DATASETS
sep.head(5)
```

```python
all = pd.concat([jan, feb, mar, apr, may, jun, jul, aug, sep],ignore_index=True)
```

```python
all
```

```python
all = all.rename(columns={"tripduration":"Duration (seconds)", 
    "start station id":"Start Station ID", 
    "start station name":"Start Station Name", 
    "start station latitude":"Latitude (start)", 
    "start station longitude":"Longitude (start)", 
    "end station id":"End Station ID",
    "end station name":"End Station Name", 
    "end station latitude":"Latitude (end)", 
    "end station longitude":"Longitude (end)",
    "bikeid":"Bike ID",
    "usertype":"Type of User",
    "birth year":"Birth Year",
    "gender":"Gender"})
all
```

```python
all['Start Date'] = pd.to_datetime(all['starttime']).dt.date
all['Start Time'] = pd.to_datetime(all['starttime']).dt.time
all['End Date'] = pd.to_datetime(all['stoptime']).dt.date
all['End Time'] = pd.to_datetime(all['stoptime']).dt.time
```

```python
all
```

```python
# DROP IRRELEVANT COLUMNS
del all['End Time']
del all['End Date']
del all['Start Time']
del all["stoptime"]
del all["starttime"]
all
```

```python
all.to_csv('citybike_jan_sept_2020')
```
